### Analysis
* As expected, the weather becomes significantly warmer as one approaches the equator (0 Deg. Latitude). More interestingly, however, is the fact that the southern hemisphere tends to be warmer this time of year than the northern hemisphere. This may be due to the tilt of the earth.
* There is no strong relationship between latitude and cloudiness. However, it is interesting to see that a strong band of cities sits at 0, 80, and 100% cloudiness.
* There is no strong relationship between latitude and wind speed. However, in northern hemispheres there is a flurry of cities with over 20 mph of wind.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
print(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it's being processed (with the city number and city name).


```python
# Generating URL for API call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
```

```python
# Creating lists to collect the desired data
city_name = []
curr_temp = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
min_temp = []
wind_speed = []

# Start call counter 
record = 1

# Log file print statement
print(f"Beginning Data Retrieval")
print(f"------------------------")

#Loop through the cities in the city list 
for city in cities:
    api_url = (f"{url}&q={city}")
    #print(api_url)
    response = requests.get(api_url).json()
    #print(response)
    
    # Since OpenWeatherMap may not have have records for all the cities generated by CityPy module
    # We use a Try statement to append data only when found
    try:
        city_name.append(response['name'])
        curr_temp.append(response['main']['temp'])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        min_temp.append(response["main"]["temp_min"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} | {city_record}")
        print(f"{url}&q={city}")
        
        # Increase counter by one 
        record= record + 1
        
        # Wait 0.5 seconds for each loop to avoid exceeding rate limit of API
        time.sleep(0.5)
        
    # If no record found "skip" to next call
    except:
        print("City not found. Skipping...")
    continue
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Create a dictonary to contain the collected data
city_weather_data = {
    "City": city_name, 
    "Country": country,
    "Lat": lat, 
    "Lng": lng,
    "Date": pd.to_datetime(date, unit='s'), # Converting epoch date to regular time stamp in UTC
    "Max Temp": max_temp,
    "Min Temp": min_temp,
    "Wind Speed":wind_speed,
    "Humidity": humidity,
    "Cloudiness":cloudiness,
}

# Create a data frame from dictionary
weather_data = pd.DataFrame(city_weather_data)

# Display count of weather data values 
weather_data.count()
```

```python
# Saving weather data
weather_data.to_csv('../output_data/weather_data.csv')

weather_data.head()
```

#### Latitude vs. Temperature Plot

```python
# Set style of scatterplot
sns.set()

# Create scatterplot of dataframe
fig = sns.regplot(x='Lat', # x axis
           y='Max Temp', # y axis
           data=weather_data, # Data source
           marker='o', color='#34495e', # Setting marker and color
           fit_reg=True, # Fix a regression line
           order=3, ci=None, truncate=True, # Set polynomial fit
           scatter_kws={"marker": "D", # Set marker style
                        "s": 25,}) # S marker size

# Set chart title
plt.title('Temperature vs City Latitude')

# Set x-axis label
plt.xlabel('Latitude')

# Set y-axis label
plt.ylabel('Maximum Temperature (F)')

plt.savefig("../figures/Max_Temp_vs_Latitude.png")
```
